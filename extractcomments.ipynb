```python
import fitz  # PyMuPDF

def extract_comments_from_pdf(pdf_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    
    comments = []

    # Iterate through each page
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        for annot in page.annots():
            # Extract the bounding box of the annotation
            bbox = annot.rect
            
            # Extract the text within the bounding box
            linked_text = page.get_textbox(bbox)
            
            # Extract the comment
            comment = {
                'page': page_num + 1,
                'linked_text': linked_text,
                'comment': annot.info.get('content', 'No content')
            }
            comments.append(comment)
    
    return comments

# Path to your PDF file
pdf_path = '/Applications/UNI/ThesisNew/Thesis_DRAFT-26_RQ.pdf'

# Extract comments
comments = extract_comments_from_pdf(pdf_path)

# Print the comments
for comment in comments:
    print(f"Page: {comment['page']}")
    print(f"Linked Text: {comment['linked_text']}")
    print(f"Comment: {comment['comment']}")
    print('---')
```

```python
import fitz  # PyMuPDF

def extract_comments_with_paragraphs_from_pdf(pdf_path, expansion_factor=100):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    
    comments = []

    # Iterate through each page
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        for annot in page.annots():
            # Extract the bounding box of the annotation
            bbox = annot.rect
            
            # Expand the bounding box
            expanded_bbox = fitz.Rect(
                bbox.x0,
                bbox.y0,
                bbox.x1 + (bbox.width * (expansion_factor - 1) / 2),
                bbox.y1
            )
            
            # Extract the text within the expanded bounding box
            linked_text = page.get_textbox(expanded_bbox)
            
            # Extract the comment
            comment = {
                'page': page_num + 1,
                'linked_text': linked_text,
                'comment': annot.info.get('content', 'No content')
            }
            comments.append(comment)
    
    return comments

# Path to your PDF file
pdf_path = '/Applications/UNI/ThesisNew/Thesis_DRAFT-26_RQ.pdf'

# Extract comments
comments = extract_comments_with_paragraphs_from_pdf(pdf_path)

# Print the comments
for comment in comments:
    print(f"Page: {comment['page']}")
    print(f"Linked Text: {comment['linked_text']}")
    print(f"Comment: {comment['comment']}")
    print('---')
```
